7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

url_mlbb = "https://raw.githubusercontent.com/mrochella/datamining/refs/heads/main/mlbb_hero.csv"
df_mlbb = pd.read_csv(url_mlbb, delimiter=';')

# Encode the 'role' column
le = LabelEncoder()
df_mlbb['role'] = le.fit_transform(df_mlbb['role'])

# Separate features and target variable
features = ['defense_overall', 'offense_overall', 'skill_effect_overall', 'difficulty_overall',
            'movement_spd', 'magic_defense', 'mana', 'hp_regen', 'physical_atk',
            'physical_defense', 'hp', 'attack_speed', 'mana_regen']
X = df_mlbb[features]
y = df_mlbb['role']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the model
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy: {accuracy}")

# Predict the role for the given character
new_character = [[6.0, 7.0, 4.0, 3.0, 250, 9, 380, 37, 119, 24, 2689, 0.75, 16.0]]
predicted_role_encoded = clf.predict(new_character)

# Get the predicted role (dicoding)
predicted_role = le.inverse_transform(predicted_role_encoded)
print(f"The predicted role for the character is: {predicted_role[0]}")
```

8. Why is CSV one of the most commonly used file formats in the field of data science? <br>

```python
# 8

# Sederhana: File CSV mudah dibuat dan dimanipulasi karena menyimpan data dalam bentuk teks biasa. Setiap editor teks atau perangkat lunak spreadsheet,
# seperti Excel atau Google Sheets, dapat membuka dan mengeditnya.

# Kompatibilitas Luas: File CSV dapat digunakan di berbagai platform dan bahasa pemrograman, menjadikannya ideal untuk pertukaran data.
# Ini termasuk kompatibilitas dengan sebagian besar alat analisis data seperti Python's Pandas, database SQL, dan sistem big data seperti Hadoop.

# Ringan: File CSV berukuran kecil dan tidak memiliki format yang rumit, sehingga ukurannya lebih kecil dan prosesnya lebih cepat dibandingkan dengan format lain,
# seperti Excel atau XML. Kesederhanaan ini juga memudahkan pemrosesan, terutama saat menangani kumpulan data besar.

# Aksesibilitas Universal: Struktur sederhana baris dan kolom dalam CSV membuatnya menjadi format yang hebat untuk penyimpanan dan berbagi data.
# Sebagian besar aplikasi modern, basis data, dan alat pemrosesan data mendukung CSV sebagai opsi impor/ekspor.

# SUMBER: https://konbert.com/blog/why-csv-is-still-king, https://www.oneschema.co/blog/csv-files, https://bytescout.com/blog/csv-format-history-advantages.html
```
